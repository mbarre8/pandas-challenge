PyCity Schools Analysis

```python
#Dependencies
import pandas as pd
```

```python
#Path to files 
School_file="Resources/schools_complete.csv"
Student_file="Resources/students_complete.csv"

#Read csv files and store in pandas dataframe
School_file_df=pd.read_csv(School_file)
Student_file_df=pd.read_csv(Student_file)

#combine the 2 files data into one dataset
Combined_schools_data_df = pd.merge(Student_file_df, School_file_df, how="left",on=["school_name","school_name"])
Combined_schools_data_df.head()
```

District Summary

```python
#used combined dataset to get total schools in district
school_count=len(Combined_schools_data_df["school_name"].unique())
print(school_count)
```

```python
#Used combined dataset to get total students in district
student_count= len(Combined_schools_data_df["student_name"])
print(student_count)
```

```python
#used schools dataframe to get total budget in district
total_budget= School_file_df["budget"].sum()
print(total_budget)
```

```python
#calculate average math score but getting mean of entire "math_score" column
average_math_score=(Combined_schools_data_df["math_score"]).mean()
print(average_math_score)
```

```python
#calculate average reading score but getting mean of entire "reading_score" column
average_reading_score=(Combined_schools_data_df["reading_score"]).mean()
print(average_reading_score)
```

```python
#used combined dataset to filter the number of students whos math scores where above 70 and divided that value by total student in district
passed_math_count=Combined_schools_data_df[(Combined_schools_data_df["math_score"]>=70)].count()["student_name"]
percent_passed_math= passed_math_count/float(student_count) * 100
percent_passed_math
```

```python
#used combined dataset to filter the number of students whos reading scores where above 70 and divided that value by total student in district
passed_reading_count=Combined_schools_data_df[(Combined_schools_data_df["reading_score"]>=70)].count()["student_name"]
percent_passed_reading= passed_reading_count/float(student_count) * 100
percent_passed_reading
```

```python
#used combined dataset to filter the number of students whos math and reading scores where above 70 and divided that value by total student in district
passed_math_reading_count=Combined_schools_data_df[(Combined_schools_data_df["math_score"]>=70) & (Combined_schools_data_df["reading_score"]>=70)].count()["student_name"]
percent_passed_overall=passed_math_reading_count/ float(student_count)*100
percent_passed_overall
```

```python
#Created dataframe with a dictionary, created column names with corresponding value containing calculations from above lines
District_Summary = pd.DataFrame({"Total Schools":[school_count],"Total Students":[student_count], "Total Budget": [total_budget], "Average Math Score": [average_math_score],"Average Reading Score":[average_reading_score],"% Passing Math":[percent_passed_math], "% Passing Reading":[percent_passed_reading], "% Overall Passing":[percent_passed_overall]})
#Formatting columns 
District_Summary["Total Students"] = District_Summary["Total Students"].map("{:,}".format)
District_Summary["Total Budget"] = District_Summary["Total Budget"].map("${:,.2f}".format)
District_Summary[["% Passing Math","% Passing Reading","% Overall Passing"]] = District_Summary[["% Passing Math","% Passing Reading","% Overall Passing"]].applymap("{:,.6f}%".format)

District_Summary
```

School Summary

```python
#school type for each school
School_Types= School_file_df.set_index(["school_name"])["type"]
School_Types
```

```python
#Use combined dataset to group by schools
Schools_df=Combined_schools_data_df.groupby(["school_name"])
#make list of all school names
##School_names=list(Combined_schools_data_df["school_name"].unique())
```

```python
#total number of students per school
#Renamed series to better reflect data presented
Total_Students_per_school= Schools_df["student_name"].count()
Total_Students_per_school.rename("Total_Students_per_school", inplace=True)
Total_Students_per_school
```

```python
#Budget per school
Budget_Per_School = School_file_df.groupby(["school_name"]).mean()["budget"]
Budget_Per_School
```

```python
#Budget allocated per student
#Renamed series to better reflect data presented
Per_School_Capita = Budget_Per_School / Total_Students_per_school
Per_School_Capita.rename("Per_school_capita", inplace=True)
Per_School_Capita
```

```python
#Calculating average math schools per school
per_school_math = Schools_df["math_score"].mean()
per_school_math
```

```python
#Calculating average reading schools per school
per_school_reading = Schools_df["reading_score"].mean()
per_school_reading
```

```python
#pulling all math scores above or equal to 70 in combined dataset
school_passing_math =Combined_schools_data_df[(Combined_schools_data_df["math_score"]>=70)]
school_passing_math
```

```python
#pulling all reading scores above or equal to 70 in combined dataset
school_passing_reading =Combined_schools_data_df[(Combined_schools_data_df["reading_score"]>=70)]
school_passing_reading
```

```python
#pulling all reading and math scores above or equal to 70 in combined dataset
passing_math_and_reading = Combined_schools_data_df[(Combined_schools_data_df["reading_score"] >= 70) & (Combined_schools_data_df["math_score"] >= 70)]
```

```python
#used combined dataset to filter the number of students whos math and reading scores where above 70 and divided that value by total student per school
#renamed all series to better reflect data
per_school_passing_math = school_passing_math.groupby(["school_name"]).count()["student_name"] / Total_Students_per_school * 100
per_school_passing_math.rename("per_school_passing_math", inplace=True)
per_school_passing_reading = school_passing_reading.groupby(["school_name"]).count()["student_name"] / Total_Students_per_school * 100
per_school_passing_reading.rename("per_school_passing_reading", inplace=True)
overall_passing_rate = passing_math_and_reading.groupby(["school_name"]).count()["student_name"] / Total_Students_per_school * 100
overall_passing_rate.rename("overall_passing_rate", inplace=True)
```

```python
#merging all calculations columns from above into one chart
#sorted values in first merge to display schools in alphabetical order

merge_df=pd.merge(School_Types, Total_Students_per_school, on="school_name").sort_values("school_name")
merge_df2=pd.merge(merge_df, Budget_Per_School, on="school_name")
merge_df3=pd.merge(merge_df2, Per_School_Capita, on="school_name")
merge_df4=pd.merge(merge_df3, per_school_math , on="school_name")
merge_df5=pd.merge(merge_df4, per_school_reading , on="school_name")
merge_df6=pd.merge(merge_df5,per_school_passing_math , on="school_name")
merge_df7=pd.merge(merge_df6,per_school_passing_reading , on="school_name")
merge_df8=pd.merge(merge_df7,overall_passing_rate, on="school_name")
```

```python
#Place data in DataFrame and renaming headers
School_Summary = pd.DataFrame({"School Type": School_Types, "Total Students": Total_Students_per_school,"Total School Budget":Budget_Per_School, "Per Student Budget":Per_School_Capita,"Average Math Score":per_school_math,"Average Reading Score": per_school_reading,"% Passing Math": per_school_passing_math, "% Passing Reading":per_school_passing_reading , "% Overall Passing":overall_passing_rate})

#formatting columns
School_Summary["Total School Budget"]= School_Summary["Total School Budget"].map("${:,.2f}".format)
School_Summary["Per Student Budget"]= School_Summary["Per Student Budget"].map("${:,.2f}".format)

#Delete index title                                                                            
School_Summary.index.name = None
School_Summary
```

Math Scores by Grade

```python
#filtering through combined dataset by grade
ninth_graders = Combined_schools_data_df[(Combined_schools_data_df["grade"] == "9th")]
tenth_graders = Combined_schools_data_df[(Combined_schools_data_df["grade"] == "10th")]
eleventh_graders = Combined_schools_data_df[(Combined_schools_data_df["grade"] == "11th")]
twelfth_graders = Combined_schools_data_df[(Combined_schools_data_df["grade"] == "12th")]
```

Scores by School Spending

```python
#Set bins and labels
spending_bins = [0, 585, 630, 645, 680]
labels = ["<$585", "$585-630", "$630-645", "$645-680"]
```

```python
#copy old dataframe into a new dataframe to pull Per student budget
school_spending_df = School_Summary.copy()
school_spending_df.dtypes
```

```python
#Create/name new column, convert Per student column to float value, based on Per student budget determine which bin each school falls in
school_spending_df["Spending Ranges (Per Student)"] = pd.cut(school_spending_df["Per Student Budget"].str.replace("$", "").astype("float"), bins=spending_bins, labels=labels)
school_spending_df
```

```python
#Calculate averages for each column
spending_math_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["Average Math Score"]
spending_reading_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["Average Reading Score"]
spending_passing_math = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["% Passing Math"]
spending_passing_reading = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["% Passing Reading"]
overall_passing_spending = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["% Overall Passing"]
```

```python
#Combine all above columns in DataFrame
spending_summary = pd.DataFrame({"Average Math Score":spending_math_scores,"Average Reading Score":spending_reading_scores,"% Passing Math":spending_passing_math,"% Passing Reading":spending_passing_reading, "% Overall Passing":overall_passing_spending})
#Results
spending_summary
```
